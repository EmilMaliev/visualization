Визуализация справа показывает количество заполненных значений в каждом столбце, что дает представление о полноте данных. Другими словами — это гистограмма полноты данных. Заменим все значения <18 и >100 лет - средним значением возраста. Для этого:

1) Сначала преобразуем эти значения (выходящие за разумные рамки) в формат **NaN** (not-a-number) - **np.nan**

2) а затем, используя **SimpleImputer**, заменим эти значения **средними значениями по колонке**:

```python
# Преобразуем значения, выходящие за разумные рамки, в формат NaN
df.loc[(df['Age'] < 18) | (df['Age'] >100), 'Age'] = np.nan
print(df["Age"].value_counts(dropna=False))
```

```python
from sklearn.impute import SimpleImputer

# Используя SimpleImputer, заменим эти значения средними значениями по колонке
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

df['Age'] = imputer.fit_transform(df[['Age']])
print(df["Age"].value_counts(dropna=False))
```

Теперь датафрейм **df** в содержит в колонке **Age**, значения в которой не
меньше 18 и не больше 72. Визуализировав значения в колонке **Age**, получим разумное распределение, показанное на графике.

```python
sns.displot(df['Age'], kde = True, aspect = 2);
plt.title('Распределение по возрасту');
```

```python
from math import ceil # Импортируем функцию ceil из модуля math для округления вверх
from textwrap import wrap # Импортируем функцию wrap из модуля textwrap для форматирования текста

# Установим стиль для графиков библиотеки seaborn
sns.set(style='whitegrid', font_scale= 1.5) # Устанавливаем стиль 'whitegrid' для фона и размер шрифта
# sns.set_palette("pastel") # Закомментированная строка для установки палитры, но мы используем 'Spectral' ниже

gr_count = len(titels_q) # Определяем количество графиков на основе длины списка заголовков
cols = 2 # Устанавливаем количество столбцов для сетки графиков
rows = ceil(gr_count/2) # Вычисляем количество строк, округляя количество графиков, деленное на 2, вверх

# Создание объектов фигуры и оси
# fig - это вся фигура, содержащая графики
# axes - это массив объектов осей (подграфиков) внутри фигуры
fig, axes = plt.subplots(ncols=cols, nrows=rows, figsize=(12 * cols, 5 * rows) ) # Создаем сетку подграфиков с заданным количеством строк и столбцов, а также общим размером фигуры

# Flatten the 2D array of axes into a 1D list
# Это необходимо, если nrows и ncols > 1, чтобы итерировать по подграфикам как по одному списку
axes = axes.flatten()

# Запускаем цикл для создания каждого графика
for i in range(0, gr_count):
    ax = axes[i] # Выбираем текущий подграфик из списка

    # Строим столбчатую диаграмму (countplot) для текущего признака (features[i])
    # data = Data: используем наш очищенный и закодированный датафрейм
    # x = features[i]: колонка, по которой будут строиться столбцы
    # ax=ax: указываем, на каком подграфике рисовать
    # order = sorted(Data[features[i]].unique()): сортируем категории для более упорядоченного отображения
    # palette='Spectral': устанавливаем цветовую палитру для столбцов
    sns.countplot(data = Data, x = features[i], ax=ax, order = sorted(Data[features[i]].unique()), palette='Spectral', hue=features[i], legend=False)
    # Добавляем метки (значения) над каждым столбцом, размещая их по центру
    for container in ax.containers:
        ax.bar_label(container, label_type='center')

    # Форматируем заголовок графика, оборачивая его, если он слишком длинный
    title = '\n'.join(wrap(titels_q[i],60))
    ax.set_title(title, fontsize=20) # Устанавливаем заголовок для текущего подграфика с заданным размером шрифта

    # https://matplotlib.org/stable/users/explain/quick_start.html
    # ax.set_xlabel("") # Закомментированная строка: можно было бы убрать метки оси X
    ax.set_ylabel("") # Убираем метки оси Y, так как они не всегда нужны для countplot

# Автоматическая регулировка расстояния между графиками
# Это предотвращает наложение заголовков или меток и улучшает читаемость
plt.tight_layout()
plt.show() # Отображаем все созданные графики
```
