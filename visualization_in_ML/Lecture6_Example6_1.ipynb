```python
import numpy as np
import pandas as pd

# Генерация синтетических данных
np.random.seed(42)

# Апельсины
oranges_weight = np.random.normal(160, 15, 50) # генерирует 50 случайных чисел, распределённых по нормальному закону с средним значением 160 и стандартным отклонением 15.
oranges_diameter = np.random.normal(7, 0.5, 50)
oranges_color_index = np.random.normal(70, 5, 50)

# Яблоки
apples_weight = np.random.normal(140, 20, 50)
apples_diameter = np.random.normal(6, 0.5, 50)
apples_color_index = np.random.normal(60, 5, 50)

# Груши
pears_weight = np.random.normal(180, 15, 50)
pears_diameter = np.random.normal(7.5, 0.7, 50)
pears_color_index = np.random.normal(40, 5, 50)
```

```python
# Формирование DataFrame
# Апельсины
oranges = pd.DataFrame({
    'weight': oranges_weight,
    'diameter': oranges_diameter,
    'color_index': oranges_color_index,
    'label': 'orange'
})
oranges
```

```python
# Яблоки
apples = pd.DataFrame({
    'weight': apples_weight,
    'diameter': apples_diameter,
    'color_index': apples_color_index,
    'label': 'apple'
})
apples.head()
```

```python
# Груши
pears = pd.DataFrame({
    'weight': pears_weight,
    'diameter': pears_diameter,
    'color_index': pears_color_index,
    'label': 'pear'
})
pears.head()
```

```python
# Объединяем данные
data = pd.concat([oranges, apples, pears], ignore_index=True)
data
```

```python
# Преобразуем метки в числовой формат
data['label'] = data['label'].map({'orange': 0, 'apple': 1, 'pear': 2})
data
```

```python
data.info()
```

```python
# Визуализация данных

import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data.replace({'label': {0: 'Orange', 1: 'Apple', 2: 'Pear'}}), hue='label')
plt.show()
```

```python
# Разделение на признаки и метки
X = data[['weight', 'diameter', 'color_index']]
y = data['label']
```

```python
# Разделение на обучающую и тестовую выборки (70:30)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Стандартизация данных

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Обучение модели k-NN

from sklearn.neighbors import KNeighborsClassifier

k = 12
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(X_train, y_train)
```

```python
# Предсказание на тестовых данных
y_pred = knn.predict(X_test)
```

```python
# Оценка модели

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print(f"Точность (accuracy): {accuracy:.2f}")
print("\nОтчет о классификации:")
print(classification_report(y_test, y_pred, target_names=['Orange', 'Apple', 'Pear']))
```

Построим **ROC-кривую**

```python
from sklearn.metrics import roc_curve, auc

# Предсказание вероятностей для тестовой выборки
y_probs = knn.predict_proba(X_test)

# Преобразуем вероятности в один столбец для ROC-кривой
y_probs_multiclass = pd.DataFrame(y_probs, columns=knn.classes_)

# Вычисляем ROC-кривую и AUC для каждого класса
fpr = {}
tpr = {}
roc_auc = {}
for i in range(len(knn.classes_)):
    fpr[i], tpr[i], _ = roc_curve(y_test == i, y_probs_multiclass[i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Построение ROC-кривой
plt.figure()
colors = ['blue', 'green', 'red']
for i in range(len(knn.classes_)):
    plt.plot(fpr[i], tpr[i], color=colors[i], lw=2, label=f'Class {knn.classes_[i]} (area = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая для каждого класса')
plt.legend(loc='lower right')
plt.show()
```

**Построение дерева решений**

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Создание и обучение модели дерева решений
clf = DecisionTreeClassifier(random_state=42, max_depth=3)
clf.fit(X_train, y_train)
```

```python
# Оценка модели на тестовой выборке
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
print(f"Точность (accuracy): {accuracy:.2f}")

print("\nОтчет о классификации:")
# Вывод отчета классификации
print(classification_report(y_test, y_pred))
```

```python
# Визуализация дерева решений
plt.figure(figsize=(12, 8))
tree.plot_tree(clf,
               feature_names=['weight', 'diameter', 'color_index'],
               class_names=['orange', 'apple', 'pear'],
               filled=True)
plt.show()
```

Каждый квадрат (**узел**) в дереве решений представляет собой **правило принятия решения**, основанное на определенной характеристике.

Параметры в каждом узле:

**feature <= threshold** (например, color_index <= 50.85): Это условие разделения. Если значение указанной характеристики (feature) для образца меньше или равно пороговому значению (threshold), то образец переходит в левый дочерний узел; в противном случае — в правый дочерний узел.

**gini**: Это коэффициент Джини, который измеряет "чистоту" узла. Он показывает, насколько однородны классы в данном узле. Значение gini от 0 (все образцы в узле принадлежат одному классу, узел "чист") до максимального значения (например, 0.667 для трех классов, когда все классы представлены поровну, узел "максимально неоднороден"). Чем ниже gini, тем "чище" узел.

**samples**: Количество обучающих образцов, которые попали в данный узел. Это общее число точек данных, которые удовлетворяют условиям всех родительских узлов до текущего.

**value** (например, [47, 36, 22]): Это распределение образцов по классам в данном узле. Например, [47, 36, 22] означает, что в этом узле находится 47 образцов класса 0 (Orange), 36 образцов класса 1 (Apple) и 22 образца класса 2 (Pear).

**class** (например, orange): Это предсказанный класс для данного узла. Он определяется как класс большинства среди образцов, находящихся в этом узле.

Мы видим, что обе модели дали одинаковую точность.
