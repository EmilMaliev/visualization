```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Сгенерируем данные о продажах пяти видов овощей
vegetables = ['Морковь', 'Картофель', 'Огурцы', 'Помидоры', 'Перец']
sales_veg = np.random.randint(150, 600, size=len(vegetables)) # Случайные продажи от 150 до 600

df_veg_sales = pd.DataFrame({
    'Овощ': vegetables,
    'Продажи': sales_veg
})

print("Данные о продажах овощей:")
print(df_veg_sales)

# 2. Построим линейный график
plt.figure(figsize=(10, 5))
plt.plot(df_veg_sales['Овощ'], df_veg_sales['Продажи'], marker='o', linestyle='-', color='purple')
plt.title('Продажи овощей (Линейный график)')
plt.xlabel('Овощ')
plt.ylabel('Продажи')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Построим столбчатую диаграмму
plt.figure(figsize=(10, 5))
plt.bar(df_veg_sales['Овощ'], df_veg_sales['Продажи'], color='orange')
plt.title('Продажи овощей (Столбчатая диаграмма)')
plt.xlabel('Овощ')
plt.ylabel('Продажи')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Генерация данных для тепловой карты
np.random.seed(42)
months = [f'Месяц {i+1}' for i in range(12)]
products = [f'Продукт {chr(65+i)}' for i in range(10)] # 10 продуктов
sales_data = np.random.randint(50, 500, size=(len(products), len(months)))
df_sales = pd.DataFrame(sales_data, index=products, columns=months)

print("Сгенерированные данные о продажах (первые 5 строк):")
print(df_sales.head())

# --- Пример 'неудачной' визуализации (Тепловая карта) ---
plt.figure(figsize=(8, 6)) # Недостаточный размер для такого количества данных
sns.heatmap(df_sales, cmap='jet', annot=True, fmt='d', cbar=True)
plt.title('Неудачная тепловая карта продаж (перегруженная, плохая цветовая палитра)')
plt.xlabel('Месяцы')
plt.ylabel('Продукты')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# --- Пример 'удачной' визуализации (Тепловая карта) ---
# Улучшения:
# 1. Меньше данных (если исходные данные слишком велики), или фокусировка на подмножестве.
#    Для демонстрации, давайте возьмем 5 продуктов вместо 10.
# 2. Использование перцептивно однородной цветовой палитры (например, 'viridis', 'plasma', 'cividis').
# 3. Адекватный размер фигуры.
# 4. Отключение аннотаций, если они делают карту слишком перегруженной, или их форматирование.
# 5. Более читаемые метки и заголовок.

# Возьмем подмножество данных для лучшей читаемости
df_sales_improved = df_sales.iloc[:5, :]

plt.figure(figsize=(12, 8)) # Достаточный размер
sns.heatmap(df_sales_improved, cmap='viridis', annot=True, fmt='d', cbar=True, linewidths=.5, linecolor='black')
plt.title('Удачная тепловая карта продаж (читаемая, подходящая цветовая палитра)')
plt.xlabel('Месяцы')
plt.ylabel('Продукты')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Генерация синтетических данных о продажах для 10 продуктов в 8 регионах
np.random.seed(42)
products_clust = [f'Продукт {chr(65+i)}' for i in range(10)]
regions = [f'Регион {i+1}' for i in range(8)]
sales_data_clust = np.random.rand(len(products_clust), len(regions)) * 1000 # Продажи от 0 до 1000

# Добавим некоторые паттерны для кластеризации
# Например, Продукты A, B, C хорошо продаются в Регионах 1, 2, 3
sales_data_clust[0:3, 0:3] += 500 # Увеличение продаж
# Продукты X, Y, Z плохо продаются везде
sales_data_clust[7:, :] *= 0.5 # Уменьшение продаж

df_sales_clust = pd.DataFrame(sales_data_clust, index=products_clust, columns=regions)

print("Сгенерированные данные о продажах (первые 5 строк, без кластеризации):")
print(df_sales_clust.head())

# Построение тепловой карты с кластеризацией
# clustermap автоматически применяет иерархическую кластеризацию к строкам и столбцам
# и переупорядочивает их на основе схожести.
plt.figure(figsize=(12, 10)) # Размер фигуры будет регулироваться clustermap
sns.clustermap(df_sales_clust, cmap='viridis', annot=False, fmt=".0f",
               linewidths=.5, linecolor='black', figsize=(12,10))
plt.suptitle('Тепловая карта продаж с кластеризацией (Продукты vs Регионы)', y=1.02, fontsize=16)
plt.show()

print("\nНа тепловой карте с кластеризацией строки и столбцы перегруппированы. Дендрограммы слева и сверху показывают иерархию кластеров. Вы можете видеть, как продукты и регионы, имеющие схожие паттерны продаж, сгруппированы вместе.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Реальные названия продуктов
products_real = [
    'Смартфон',
    'Ноутбук',
    'Планшет',
    'Умные часы',
    'Наушники',
    'Телевизор',
    'Игровая приставка',
    'Кофемашина'
]

# Регионы продаж
regions_real = ['Северный регион', 'Южный регион', 'Восточный регион', 'Западный регион']

# Генерация синтетических данных о продажах
# Зададим seed для воспроизводимости
np.random.seed(42)

sales_data_real = np.random.randint(100, 1000, size=(len(products_real), len(regions_real)))

# Добавим некоторые 'реалистичные' паттерны для разнообразия:
# Например, смартфоны и ноутбуки хорошо продаются везде
sales_data_real[0, :] = np.random.randint(800, 1500, size=len(regions_real))
sales_data_real[1, :] = np.random.randint(700, 1200, size=len(regions_real))

# Кофемашины могут иметь более низкие продажи в одном регионе
sales_data_real[7, 0] = np.random.randint(50, 200) # Низкие продажи в Северном

df_real_sales = pd.DataFrame(sales_data_real, index=products_real, columns=regions_real)

print("Сгенерированные данные о продажах реальных продуктов:")
print(df_real_sales)

# Построение тепловой карты с пастельными цветами
plt.figure(figsize=(10, 8))

# Использование пастельной цветовой палитры. 'YlGnBu' - хороший выбор для пастельных тонов.
# annot=True для отображения значений, fmt='d' для целочисленных значений.
sns.heatmap(df_real_sales, cmap='YlGnBu', annot=True, fmt='d', cbar=True,
            linewidths=.5, linecolor='black')

plt.title('Продажи электроники по регионам (Тепловая карта)', fontsize=16)
plt.xlabel('Регион продаж', fontsize=12)
plt.ylabel('Продукт', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Добавление кластерной тепловой карты
print("\nТепловая карта продаж с кластеризацией для реальных продуктов:")
plt.figure(figsize=(12, 10)) # Размер фигуры будет регулироваться clustermap
sns.clustermap(df_real_sales, cmap='YlGnBu', annot=True, fmt='d',
               linewidths=.5, linecolor='black', figsize=(12,10))
plt.suptitle('Кластерная тепловая карта продаж электроники по регионам', y=1.02, fontsize=16)
plt.show()

print("\nНа кластерной тепловой карте вы можете увидеть, как продукты и регионы с похожими паттернами продаж сгруппированы вместе, что позволяет легче выявлять тенденции и взаимосвязи.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных Титаника (если еще не загружены)
try:
    titanic = sns.load_dataset('titanic')
except Exception as e:
    print(f"Ошибка при загрузке данных: {e}. Попробуйте загрузить из другого источника.")
    # Если загрузка не удалась, можно попробовать загрузить из файла
    # titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Удаляем пропущенные значения в колонке 'age', так как они не могут быть отображены
titanic_cleaned = titanic.dropna(subset=['age'])

# Определяем возрастные группы (bins)
bins = np.arange(0, 81, 5) # От 0 до 80 лет с шагом 5 лет

# Группируем данные по полу и возрастным группам
female_ages = pd.cut(titanic_cleaned[titanic_cleaned['sex'] == 'female']['age'], bins=bins, right=False)
male_ages = pd.cut(titanic_cleaned[titanic_cleaned['sex'] == 'male']['age'], bins=bins, right=False)

female_counts = female_ages.value_counts().sort_index()
male_counts = male_ages.value_counts().sort_index()

# Для пирамиды мужские значения часто отображают как отрицательные
male_counts_neg = -male_counts

# Убедимся, что индексы совпадают для обоих серий (чтобы правильно сопоставить возрастные группы)
combined_index = female_counts.index.union(male_counts.index)
female_counts = female_counts.reindex(combined_index, fill_value=0)
male_counts_neg = male_counts_neg.reindex(combined_index, fill_value=0)

# Построение пирамиды
fig, ax = plt.subplots(figsize=(12, 8))

# Построение баров для женщин и мужчин с более спокойными цветами
ax.barh(female_counts.index.astype(str), female_counts, color='lightcoral', label='Женщины') # Светлый коралловый
ax.barh(male_counts_neg.index.astype(str), male_counts_neg, color='lightskyblue', label='Мужчины') # Светло-голубой

# Добавление надписей к барам
# Для женщин
for i, (age_group, count) in enumerate(female_counts.items()):
    if count > 0:
        ax.text(count + 5, i, str(count), va='center', ha='left', fontsize=9, color='black')

# Для мужчин
for i, (age_group, count) in enumerate(male_counts_neg.items()):
    if count < 0:
        ax.text(count - 5, i, str(abs(count)), va='center', ha='right', fontsize=9, color='black')

# Настройка осей и заголовков
ax.set_title('Возрастная пирамида пассажиров Титаника')
ax.set_xlabel('Количество пассажиров')
ax.set_ylabel('Возрастные группы')

# Изменяем метки оси X, чтобы отрицательные значения отображались как положительные
# и располагались по центру.
# Находим максимальное абсолютное значение для установки симметричных пределов.
max_abs_val = max(abs(male_counts_neg.min()), female_counts.max())
ax.set_xlim(-max_abs_val - 20, max_abs_val + 20) # Увеличим пределы для надписей

ticks = ax.get_xticks()
ax.set_xticklabels([int(abs(tick)) for tick in ticks])

ax.legend()
ax.grid(axis='x', alpha=0.75)
plt.tight_layout()
plt.show()
```
