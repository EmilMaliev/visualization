```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Сгенерируем данные
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
data = {
    'Date': dates,
    'Value_A': np.random.randn(100).cumsum() + 50,
    'Value_B': np.random.randn(100).cumsum() + 30
}
df = pd.DataFrame(data)

print("Сгенерированные данные (первые 5 строк):")
print(df.head())

# 2. Построим Линейный график (Line Plot)
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Value_A'], label='Value A', marker='o', markersize=4, linestyle='-')
plt.plot(df['Date'], df['Value_B'], label='Value B', marker='x', markersize=4, linestyle='--')
plt.title('Линейный график')
plt.xlabel('Дата')
plt.ylabel('Значение')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Построим Диаграмму с областями (Area Chart)
plt.figure(figsize=(12, 6))
plt.fill_between(df['Date'], df['Value_A'], color='skyblue', alpha=0.4, label='Value A')
plt.fill_between(df['Date'], df['Value_B'], color='lightcoral', alpha=0.4, label='Value B')
# Для стековой диаграммы с областями можно использовать:
# plt.stackplot(df['Date'], df['Value_A'], df['Value_B'], labels=['Value A', 'Value B'], alpha=0.8)

plt.title('Диаграмма с областями')
plt.xlabel('Дата')
plt.ylabel('Значение')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Сгенерируем данные для столбчатых диаграмм
categories = ['Продукт A', 'Продукт B', 'Продукт C', 'Продукт D', 'Продукт E']
values = np.random.randint(10, 100, size=len(categories))
df_bar_new = pd.DataFrame({'Категория': categories, 'Значение': values})

print("Сгенерированные данные для столбчатых диаграмм:")
print(df_bar_new.head())

# 2. Построим Вертикальную столбчатую диаграмму (Bar Chart) с разными цветами и подписями
plt.figure(figsize=(10, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightpink']
bars = plt.bar(df_bar_new['Категория'], df_bar_new['Значение'], color=colors)

plt.title('Вертикальная столбчатая диаграмма')
plt.xlabel('Категория')
plt.ylabel('Значение')
plt.grid(axis='y', linestyle='--')

# Добавление подписей к столбцам
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, round(yval, 2), ha='center', va='bottom') # +0.5 для небольшого отступа над столбцом

plt.tight_layout()
plt.show()

# 3. Построим Горизонтальную столбчатую диаграмму (Horizontal Bar Chart) с разными цветами и подписями
plt.figure(figsize=(10, 6))
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightpink'] # Можно использовать те же или другие цвета
bars = plt.barh(df_bar_new['Категория'], df_bar_new['Значение'], color=colors)

plt.title('Горизонтальная столбчатая диаграмма')
plt.xlabel('Значение')
plt.ylabel('Категория')
plt.grid(axis='x', linestyle='--')

# Добавление подписей к столбцам
for bar in bars:
    xval = bar.get_width()
    plt.text(xval + 0.5, bar.get_y() + bar.get_height()/2, round(xval, 2), va='center') # +0.5 для небольшого отступа справа

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate data for demonstration
np.random.seed(42)
categories = ['Отдел A', 'Отдел B', 'Отдел C']
subcategories = ['Продукт X', 'Продукт Y', 'Продукт Z']
years = [2021, 2022, 2023]

data = []
for year in years:
    for category in categories:
        for subcategory in subcategories:
            value = np.random.randint(50, 200)
            data.append([year, category, subcategory, value])

df_complex_bar = pd.DataFrame(data, columns=['Год', 'Отдел', 'Продукт', 'Выручка'])

print("Сгенерированные данные для демонстрации (первые 5 строк):")
print(df_complex_bar.head())
```

### Сложный график: Накопительная столбчатая диаграмма (Stacked Bar Chart)

Этот график пытается показать выручку по продуктам и отделам за несколько лет на одной диаграмме. Он может быть информативным, но также может быть перегруженным и трудным для восприятия, если категорий много.

```python
import seaborn as sns

# 2. "Сложный" график - Накопительная столбчатая диаграмма
plt.figure(figsize=(12, 7))
sns.barplot(x='Год', y='Выручка', hue='Продукт', data=df_complex_bar, errorbar=None, dodge=False)
plt.title('Общая выручка по годам, отделам и продуктам (сложный график)')
plt.xlabel('Год')
plt.ylabel('Выручка')
plt.xticks(rotation=45)
plt.legend(title='Продукт')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

### Простые графики: Раздельные столбчатые диаграммы

Теперь давайте разделим эту информацию на несколько более простых и легко читаемых графиков. Каждый график будет сфокусирован на одной конкретной детали.

```python
# 3. Несколько "простых" графиков

# График 1: Общая выручка по отделам за все годы
plt.figure(figsize=(8, 5))
sns.barplot(x='Отдел', y='Выручка', data=df_complex_bar.groupby('Отдел')['Выручка'].sum().reset_index(), palette='viridis')
plt.title('Общая выручка по отделам за 2021-2023 гг.')
plt.xlabel('Отдел')
plt.ylabel('Общая выручка')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

# График 2: Выручка по продуктам за 2023 год
plt.figure(figsize=(8, 5))
df_2023 = df_complex_bar[df_complex_bar['Год'] == 2023].groupby('Продукт')['Выручка'].sum().reset_index()
sns.barplot(x='Продукт', y='Выручка', data=df_2023, palette='magma')
plt.title('Выручка по продуктам за 2023 год')
plt.xlabel('Продукт')
plt.ylabel('Выручка')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

# График 3: Изменение выручки Отдела B по годам
plt.figure(figsize=(8, 5))
df_dept_B = df_complex_bar[df_complex_bar['Отдел'] == 'Отдел B'].groupby('Год')['Выручка'].sum().reset_index()
sns.barplot(x='Год', y='Выручка', data=df_dept_B, palette='cividis')
plt.title('Изменение выручки Отдела B по годам')
plt.xlabel('Год')
plt.ylabel('Выручка')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

## Сводка по сравнению Bar Plot'ов:

Мы продемонстрировали построение столбчатых диаграмм одного и того же набора данных с использованием трех различных подходов:

1.  **`pandas.plot()` (Просто)**: Отлично подходит для быстрого исследовательского анализа. С минимальным кодом позволяет получить базовую визуализацию прямо из DataFrame.
2.  **Seaborn (Красиво)**: Обеспечивает эстетически приятные графики с хорошими настройками по умолчанию, часто с одной строкой кода. Идеально подходит, когда нужна более привлекательная визуализация с минимальными усилиями.
3.  **Кастомизированный Matplotlib (Точно)**: Дает полный контроль над каждым элементом графика, что позволяет создавать высокоточные и индивидуальные визуализации для публикаций или отчетов. Требует большего количества кода, но предлагает максимальную гибкость.

Каждый метод имеет свои преимущества в зависимости от ваших целей: быстрое исследование, презентабельный вид или максимальный контроль над деталями.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.transform import factor_cmap
from bokeh.palettes import Category10

# Configure Bokeh to output directly to the notebook
output_notebook()

# 1. Используем тот же набор данных df_bar_comparison
# df_bar_comparison = pd.DataFrame({'Категория': categories, 'Значение': values})

# 2. Создаем ColumnDataSource из DataFrame Bokeh, который нужен для интерактивности
source = ColumnDataSource(df_bar_comparison)

# 3. Определяем категории для оси X
categories = source.data['Категория'].tolist()

# 4. Создаем объект Figure
p = figure(x_range=FactorRange(factors=categories),
           height=350,
           title="Интерактивная столбчатая диаграмма с Bokeh",
           x_axis_label='Категория Продукта',
           y_axis_label='Общая Выручка',
           tools="pan,wheel_zoom,box_zoom,reset,save,hover")

# 5. Добавляем вертикальные столбцы
p.vbar(x='Категория', top='Значение', width=0.9,
       source=source,
       line_color='white',
       fill_color=factor_cmap('Категория', palette=Category10[len(categories)], factors=categories))

# 6. Настраиваем оси и сетку
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.major_label_orientation = 0.8 # Поворот меток оси X

# 7. Отображаем график
show(p)
```

## Сводка по интерактивным графикам:

Мы построили интерактивные столбчатые диаграммы, используя три разные библиотеки, каждая из которых имеет свои особенности:

1.  **Plotly Express**: Очень прост в использовании, часто требует всего одну-две строки кода для создания красивых и интерактивных графиков. Идеально подходит для быстрого исследования и презентаций.
2.  **Bokeh**: Предоставляет больше контроля над деталями интерактивности и внешним видом, чем Plotly Express, но требует немного больше кода. Отлично подходит для создания дашбордов и веб-приложений.
3.  **Altair**: Основан на декларативном синтаксисе (Vega-Lite), что делает его мощным для сложных статистических визуализаций. Может быть очень лаконичным, но имеет свой уникальный подход к построению графиков.

```python
import folium
import numpy as np
import pandas as pd

# Создаем базовую карту Folium, центрированную на определенных координатах (теперь Томск)
m = folium.Map(location=[56.4977, 84.9744], zoom_start=9)

# --- Генерируем данные для точек продаж ---
np.random.seed(42) # Для воспроизводимости
num_sales_points = 20

# Координаты для примера (вокруг Томска)
base_lat = 56.4977
base_lon = 84.9744

sales_data = {
    'Lat': base_lat + np.random.uniform(-0.5, 0.5, num_sales_points),
    'Lon': base_lon + np.random.uniform(-1, 1, num_sales_points),
    'Sales': np.random.randint(1000, 10000, num_sales_points),
    'Location': [f'Точка продаж {i+1}' for i in range(num_sales_points)]
}
df_sales_points = pd.DataFrame(sales_data)

print("Сгенерированные точки продаж (первые 5 строк):")
print(df_sales_points.head())

# Добавляем маркеры для каждой точки продаж на карту
for idx, row in df_sales_points.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Lon']],
        popup=f"<b>{row['Location']}</b><br>Продажи: {row['Sales']:,}",
        tooltip=f"{row['Location']} ({row['Sales']:,})",
        icon=folium.Icon(color='green', icon='shopping-cart') # Изменим иконку и цвет
    ).add_to(m)

# Добавляем центральный маркер (Центр Томска) как ориентир
folium.Marker(
    location=[56.4977, 84.9744],
    popup='<b>Центр Томска</b>',
    tooltip='Томск',
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(m)

# Отображаем карту
m
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

print("\n--- Пример социальной сети с 30 людьми и разноцветными связями (NetworkX) ---")

# 1. Создаем пустой граф
G = nx.Graph()

# 2. Генерируем 30 уникальных имен для пользователей
# Используем небольшой список имен для разнообразия
first_names = ['Иван', 'Мария', 'Петр', 'Анна', 'Дмитрий', 'Елена', 'Сергей', 'Ольга', 'Андрей', 'Наталья']
last_names = ['Смирнов', 'Иванова', 'Петров', 'Кузнецова', 'Соколов', 'Михайлова', 'Федоров', 'Морозова', 'Волков', 'Лебедева']

users = []
while len(users) < 30:
    name = f"{random.choice(first_names)} {random.choice(last_names)}"
    if name not in users:
        users.append(name)

G.add_nodes_from(users)

# 3. Добавляем ребра (случайные связи между пользователями)
# Создадим случайное количество связей (например, от 1 до 5) для каждого пользователя
# Собираем цвета для ребер
edge_colors = []
color_palette = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan', 'magenta']

for u in users:
    num_connections = random.randint(1, 5) # Каждый пользователь имеет от 1 до 5 связей
    potential_friends = [v for v in users if v != u and not G.has_edge(u, v)]

    # Если потенциальных друзей меньше, чем нужно для num_connections,
    # возьмем всех доступных
    friends_to_add = random.sample(potential_friends, min(num_connections, len(potential_friends)))

    for f in friends_to_add:
        G.add_edge(u, f)
        # Присваиваем случайный цвет ребру
        edge_colors.append(random.choice(color_palette))

# 4. Выводим информацию о графе
print(f"Количество пользователей (узлов) в сети: {G.number_of_nodes()}")
print(f"Количество связей (ребер) в сети: {G.number_of_edges()}")

# 5. Визуализируем граф социальной сети
plt.figure(figsize=(15, 12))
# pos - это функция, которая определяет расположение узлов
# spring_layout - часто используется для создания 'пружинного' расположения
pos = nx.spring_layout(G, seed=42, k=0.5, iterations=50) # Увеличим k для лучшего разделения, больше итераций

nx.draw_networkx(
    G,
    pos=pos,
    with_labels=True,
    node_color='lightgreen', # Цвет узлов
    node_size=800, # Размер узлов
    edge_color=edge_colors, # Используем список цветов для ребер
    linewidths=0.5,
    font_size=7, # Размер шрифта для меток
    font_weight='normal',
    alpha=0.8 # Прозрачность для лучшей видимости
)

plt.title('Социальная сеть (30 пользователей) с NetworkX: Разноцветные связи', fontsize=20)
plt.axis('off') # Отключаем оси
plt.tight_layout()
plt.show()
```

## Сводка по графику Pygal:

Мы создали интерактивный столбчатый график с использованием библиотеки Pygal. Pygal известен тем, что генерирует красивые, интерактивные графики в формате SVG (Scalable Vector Graphics), которые хорошо масштабируются и могут быть легко встроены в веб-страницы.

*   **Интерактивность**: График Pygal, отображаемый в ноутбуке, поддерживает интерактивность, такую как всплывающие подсказки при наведении на столбцы.
*   **Формат SVG**: Преимущество SVG в том, что графики выглядят четко при любом масштабировании, без потери качества.
*   **Простота использования**: Для создания базового графика требуется минимальное количество кода.

Pygal отлично подходит, когда вам нужны высококачественные, масштабируемые и интерактивные графики, особенно для веб-приложений.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("\n--- Визуализация для Insight: Выявление аномалий в производительности ---")

# 1. Сгенерируем данные: Производительность отделов
np.random.seed(42)
offices = ['Отдел A', 'Отдел B', 'Отдел C', 'Отдел D', 'Отдел E', 'Отдел F', 'Отдел G']
performance = np.random.randint(70, 95, size=len(offices)) # Производительность от 70 до 95

# Создадим один отдел с заметно низкой производительностью (аномалия)
performance[3] = np.random.randint(30, 50) # Отдел D имеет низкую производительность

df_performance = pd.DataFrame({'Отдел': offices, 'Производительность': performance})

print("Сгенерированные данные о производительности (первые 5 строк):")
print(df_performance.head())

# 2. Построим столбчатую диаграмму для визуализации производительности
plt.figure(figsize=(10, 6))
bars = plt.bar(df_performance['Отдел'], df_performance['Производительность'], color='skyblue')

# Окрасим выделяющийся отдел другим цветом для усиления insight
colors = ['skyblue'] * len(offices)
colors[3] = 'red' # Отдел D окрасим в красный
for bar, color in zip(bars, colors):
    bar.set_color(color)

plt.title('Производительность отделов', fontsize=16)
plt.xlabel('Отдел', fontsize=12)
plt.ylabel('Оценка производительности', fontsize=12)
plt.ylim(0, 100) # Установим одинаковую шкалу для лучшего сравнения
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Добавим подписи значений над столбцами
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, round(yval, 0), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 3. Объяснение Insight
print("\n--- Insight, полученный из визуализации ---")
print("На этом графике сразу видно, что \"Отдел D\" имеет значительно более низкую оценку производительности по сравнению с другими отделами. Это \"insight\".")
print("Без визуализации, просматривая просто список чисел, эту аномалию можно было бы легко пропустить или недооценить её значимость.")
print("Визуализация мгновенно привлекает внимание к \"Отделу D\" (дополнительно выделенному красным цветом), указывая на необходимость дальнейшего расследования причин низкой производительности именно в этом отделе.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("\n--- 3D график для Insight: Поиск оптимальной ценовой стратегии ---")

# 1. Сгенерируем данные для 3D
np.random.seed(42) # Для воспроизводимости
num_products = 100

# X: Цена продукта (от 10 до 200)
product_prices = np.random.uniform(10, 200, num_products)

# Y: Объем продаж (чем выше цена, тем обычно ниже объем, но с шумом)
sales_volume = 1000 - 4 * product_prices + np.random.normal(loc=0, scale=80, size=num_products)
sales_volume[sales_volume < 50] = 50 # Минимальный объем продаж

# Z: Прибыль (может зависеть от цены и объема, например, оптимальная прибыль в среднем ценовом сегменте)
profit_margin = 0.5 - (product_prices - 100)**2 / 20000 + np.random.normal(loc=0, scale=0.05, size=num_products)
profit_margin[profit_margin < 0.1] = 0.1 # Минимальная маржа
profit_margin[profit_margin > 0.6] = 0.6 # Максимальная маржа

# Для демонстрации 'инсайта', создадим один продукт-выброс:
# Продукт с высокой ценой, неплохим объемом, но очень низкой прибылью (возможно, из-за высоких издержек)
product_prices[5] = 180 # Высокая цена
sales_volume[5] = 400  # Средний объем
profit_margin[5] = 0.05 # Очень низкая маржа - потенциальный "плохой" продукт

# 2. Создаем 3D график
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 3. Добавляем точки на 3D график
# Используем цвет для отображения прибыли (Z), а размер точки - для объема продаж (Y)
scatter = ax.scatter(product_prices, sales_volume, profit_margin,
                     c=profit_margin, cmap='viridis',
                     s=sales_volume * 0.5, # Размер точки пропорционален объему продаж
                     alpha=0.7, edgecolor='k')

# 4. Устанавливаем заголовки осей
ax.set_xlabel('Цена продукта')
ax.set_ylabel('Объем продаж')
ax.set_zlabel('Прибыль (маржа)')

# 5. Устанавливаем заголовок графика
ax.set_title('3D: Цена vs. Объем продаж vs. Прибыль', fontsize=16)

# 6. Добавляем цветовую шкалу для прибыли
cbar = fig.colorbar(scatter, shrink=0.5, aspect=10)
cbar.set_label('Маржа прибыли')

# 7. Объяснение Insight
print("\n--- Insight, полученный из 3D визуализации ---")
print("На этом 3D графике мы можем увидеть взаимосвязь трех переменных: цены продукта, объема продаж и маржи прибыли.")
print("Видно, что большинство продуктов с высокой прибылью (яркие цвета) находятся в определенном диапазоне цен и объемов продаж. Это помогает выявить \"золотую середину\" ценовой стратегии.")
print("\nОсобое внимание привлекает одна точка (продукт с высокой ценой, неплохим объемом, но очень низкой маржой прибыли - наш искусственно созданный выброс).")
print("В 2D проекциях эта аномалия могла бы быть менее заметной, но в 3D она ярко выделяется из общей картины, указывая на потенциальную проблему, требующую расследования (например, высокие производственные издержки или неэффективная логистика для этого конкретного продукта). Это и есть ключевой \"инсайт\".")

# 8. Отображаем график
plt.tight_layout()
plt.show()
```
