```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

# ЗАДАНИЕ 1.1. Для финального отчета. Вариант 16

## Постройте график функции $y(x)=(1-2x)\cos(20x)$.

```python
# отрезок и шаг
dx = 0.01
x = np.arange(0, 2 + dx, dx)

# функция
y = (1 - 2 * x) * np.cos(20 * x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$y(x)=(1-2x)\cos(20x)$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции y(x)')
plt.grid(True)
plt.legend()

plt.show()
```

## Постройте в одном окне графики двух функции $y_1(x)=10e^{0.5x}$ и $y_2(x)=\cos^2(20x)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# отрезок и шаг
dx = 0.01
x = np.arange(0, 1 + dx, dx)

# функции
y1 = 10 * np.exp(0.5 * x)
y2 = np.cos(20 * x) ** 2


plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y_1(x)=10e^{0.5x}$')
plt.plot(x, y2, label=r'$y_2(x)=\cos^2(20x)$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Графики функций $y_1(x)$ и $y_2(x)$')
plt.grid(True)
plt.legend()

plt.show()
```

## Добавьте к указанным двум графикам третий – по своему усмотрению ($y_3(x)=10x$). И постройте эти три графика в разделенных окнах:
- Используя функцию subplot();
- Используя функцию subplots();
- Свободная компоновка с GridSpec

```python
dx = 0.01
x = np.arange(0, 1 + dx, dx)

y1 = 10 * np.exp(0.5 * x)
y2 = np.cos(20 * x) ** 2
y3 = 10 * x
```

### subplot()

```python
plt.figure(figsize=(8, 8))

plt.subplot(3, 1, 1)
plt.plot(x, y1)
plt.title(r'$y_1(x)=10e^{0.5x}$')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(x, y2)
plt.title(r'$y_2(x)=\cos^2(20x)$')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(x, y3)
plt.title(r'$y_3(x)=10x$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### subplots()

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 8))

axs[0].plot(x, y1)
axs[0].set_title(r'$y_1(x)=10e^{0.5x}$')
axs[0].grid(True)

axs[1].plot(x, y2)
axs[1].set_title(r'$y_2(x)=\cos^2(20x)$')
axs[1].grid(True)

axs[2].plot(x, y3)
axs[2].set_title(r'$y_3(x)=10x$')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

### GridSpec

```python
fig = plt.figure(figsize=(8, 8))
gs = GridSpec(3, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, :])   
ax2 = fig.add_subplot(gs[1, 0])   
ax3 = fig.add_subplot(gs[1:, 1])  

ax1.plot(x, y1)
ax1.set_title(r'$y_1(x)=10e^{0.5x}$')
ax1.grid(True)

ax2.plot(x, y2)
ax2.set_title(r'$y_2(x)=\cos^2(20x)$')
ax2.grid(True)

ax3.plot(x, y3)
ax3.set_title(r'$y_3(x)=10x$')
ax3.grid(True)

plt.tight_layout()
plt.show()
```

# ЗАДАНИЕ 1.2. Для финального отчета. Вариант 16

## Постройте график функции $y = \cos(\cos x) + \frac{\sqrt{3}}{2}x - \frac{1}{2}$. График должен содержать заголовок и подписи осей.

```python
x = np.linspace(-5, 5)

# функция
y = np.cos(np.cos(x)) + (np.sqrt(3) / 2) * x - 0.5

plt.figure(figsize=(8, 5))
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$y = \cos(\cos x) + \frac{\sqrt{3}}{2}x - \frac{1}{2}$')
plt.grid(True)

plt.show()
```

##  Скопируйте код для построения графика и измените внешний вид заголовка и подписей осей, используя необходимые параметры.

```python
x = np.linspace(-5, 5, 1000)

# функция
y = np.cos(np.cos(x)) + (np.sqrt(3) / 2) * x - 0.5

plt.figure(figsize=(9, 5))
plt.plot(x, y, linewidth=2, color='darkblue')

plt.xlabel('Переменная x', fontsize=12, fontweight='bold', color='black', labelpad=10)
plt.ylabel('Значение y', fontsize=12, fontweight='bold', color='black', labelpad=10)

plt.title(
    r'График функции $y = \cos(\cos x) + \frac{\sqrt{3}}{2}x - \frac{1}{2}$',
    fontsize=14,
    fontweight='bold',
    color='darkred',
    pad=15
)

plt.grid(True)
plt.show()
```

## Используя аннотацию – выделите: локальный минимум (или максимум) функции, точку разрыва, точку перегиба. Используйте разные типы стрелок.

```python
x = np.linspace(-5, 5, 2000)
y = np.cos(np.cos(x)) + (np.sqrt(3) / 2) * x - 0.5

# первая и вторая производные (численно)
dy = np.gradient(y, x)
d2y = np.gradient(dy, x)

# локальный минимум (пример)
idx_min = np.argmin(y)
x_min, y_min = x[idx_min], y[idx_min]

# локальный максимум
idx_max = np.argmax(y)
x_max, y_max = x[idx_max], y[idx_max]

# точка перегиба (смена знака второй производной)
infl_idx = np.where(np.diff(np.sign(d2y)))[0][2]
x_inf, y_inf = x[infl_idx], y[infl_idx]

plt.figure(figsize=(9, 5))
plt.plot(x, y, linewidth=2, color='navy')

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title(
    r'Аннотации: экстремумы и перегиб',
    fontsize=14, fontweight='bold'
)
plt.grid(True)

#  локальный минимум — красный треугольник вниз
plt.scatter(
    x_min, y_min,
    marker='v',
    s=200,
    color='red',
    zorder=5
)
plt.text(
    x_min, y_min + 0.8,
    'Локальный минимум',
    color='red',
    fontsize=11,
    ha='center'
)

# локальный максимум — зелёный треугольник вверх
plt.scatter(
    x_max, y_max,
    marker='^',
    s=200,
    color='green',
    zorder=5
)
plt.text(
    x_max, y_max + 0.5,
    'Локальный максимум',
    color='green',
    fontsize=11,
    ha='center'
)


# точка перегиба — как было
plt.annotate(
    'Точка перегиба',
    xy=(x_inf, y_inf),
    xytext=(x_inf + 1.5, y_inf - 1),
    arrowprops=dict(arrowstyle='fancy', color='blue'),
    fontsize=11,
    color='blue'
)
plt.annotate(
    'Точек разрыва нет\n(функция непрерывна)',
    xy=(2, y[np.abs(x - 2).argmin()]),
    xytext=(3.5, 2),
    arrowprops=dict(arrowstyle='-[', linestyle='dashed', color='purple'),
    fontsize=11,
    color='purple'
)


plt.show()
```

## Добавьте на график надпись (текст придумайте самостоятельно).

## Добавьте еще один график $y = 5x$, оформите легенду.

```python
x = np.linspace(-5, 5)

y1 = np.cos(np.cos(x)) + (np.sqrt(3) / 2) * x - 0.5
y2 = 5 * x

plt.figure(figsize=(9, 5))

plt.plot(
    x, y1,
    linewidth=2,
    color='navy',
    label=r'$y = \cos(\cos x) + \frac{\sqrt{3}}{2}x - \frac{1}{2}$'
)

plt.plot(
    x, y2,
    linestyle='--',
    color='darkorange',
    linewidth=2,
    label=r'$y = 5x$'
)

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Графики функций', fontsize=14, fontweight='bold')

plt.legend(
    fontsize=11,
    loc='upper left',
    frameon=True,
    shadow=True
)

plt.text(
    -4.5, 2.5,
    'Сравнение\nнелинейной и\nлинейной функций',
    fontsize=11,
    bbox=dict(facecolor='white', alpha=0.8)
)

plt.grid(True)
plt.show()
```
