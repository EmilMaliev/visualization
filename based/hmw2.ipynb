# ЗАДАНИЕ 2.1. Для финального отчета. Вариант 1

## Постройте в одном окне графики двух функции y1(x) и y2(x), xε[0,2π], ∆x=0,2. При построении графиков вид функций и параметры выбрать согласно варианту. Толщину линии, размер и цвет маркера, а также контура маркера подобрать самостоятельно.


```python
x = np.arange(0, 2 * np.pi, 0.2)

# функции
y1 = np.exp(-0.5 * x)
y2 = np.cos(3 * x)

plt.figure(figsize=(8, 5))

plt.plot(
    x, y1,
    color='yellow',
    linestyle='-',
    marker='.',
    linewidth=2,
    markersize=8,
    markeredgecolor='black',
    label=r'$y_1(x) = e^{-0.5x}$'
)

plt.plot(
    x, y2,
    color='pink',
    linestyle='--',
    marker='o',
    linewidth=2,
    markersize=6,
    markeredgecolor='red',
    label=r'$y_2(x) = \cos(3x)$'
)

plt.title('Графики функций $y_1(x)$ и $y_2(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.show()
```

## Подберите наилучший вариант маркировки (на основе визуализации и сравнения нескольких вариантов).

```python
x = np.linspace(0, 2*np.pi, 500)
y = np.cos(3 * x)

variants = [
    dict(title="Вариант A: 20",
         linewidth=2.5, markersize=5, markeredgewidth=0.9,
         markeredgecolor="crimson", markevery=20, alpha=1.0),

    dict(title="Вариант B: slice(50, 400, 25)",
         linewidth=2.5, markersize=5, markeredgewidth=0.9,
         markeredgecolor="crimson", markevery=slice(50, 400, 25), alpha=1.0),

    dict(title="Вариант C: [10, 50, 100]",
         linewidth=2.5, markersize=5, markeredgewidth=0.9,
         markeredgecolor="crimson", markevery=[10, 50, 100], alpha=1.0),

    dict(title="Вариант D: [0, 100, 200, 300]",
         linewidth=2.5, markersize=5, markeredgewidth=0.9,
         markeredgecolor="crimson", markevery=[0, 100, 200, 300], alpha=1.0),
]

fig, axes = plt.subplots(2, 2, figsize=(11, 7), constrained_layout=True)

for ax, v in zip(axes.ravel(), variants):
    ax.plot(
        x, y,
        color="pink",
        linestyle="--",
        marker="o",
        linewidth=v["linewidth"],
        markersize=v["markersize"],
        markeredgecolor=v["markeredgecolor"],
        markeredgewidth=v["markeredgewidth"],
        markevery=v["markevery"],
        alpha=v["alpha"],
    )
    ax.set_title(v["title"])
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True)

plt.show()
```

Науилучиший вариант А, на основе визуализации самый читаемый график.

## Продемонстрируйте навыки построения графиков с заливкой (2-3 варианта)

```python
x = np.linspace(0, 2*np.pi, 500)
y = np.cos(3 * x)

plt.figure(figsize=(8, 5))

plt.plot(x, y, color='pink', linestyle='--', linewidth=2, label=r'$y=\cos(3x)$')
plt.fill_between(x, y, 0, color='pink', alpha=0.3)

plt.title('Заливка под графиком $y = \cos(3x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.plot(x, y, color='pink', linestyle='--', linewidth=2)

plt.fill_between(
    x, y, 0,
    where=(y >= 0),
    interpolate=True,
    color='lightcoral',
    alpha=0.4,
    label='y ≥ 0'
)

plt.fill_between(
    x, y, 0,
    where=(y < 0),
    interpolate=True,
    color='lightblue',
    alpha=0.4,
    label='y < 0'
)

plt.title('Раздельная заливка положительных и отрицательных значений')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.show()
```

## Продемонстрируйте навыки обрезки графика (ограничения для обрезки придумайте самостоятельно), $ y <= 0.7 $, $ π/2 < x < 3π/2 $>>.

```python
x = np.linspace(0, 2*np.pi, 500)
y = np.cos(3 * x)

y_masked = np.ma.masked_where(y > 0.7, y)
plt.figure(figsize=(8, 5))

plt.plot(x, y_masked, color='pink', linestyle='--', linewidth=2)

plt.xlim(np.pi/2, 3*np.pi/2)

plt.title('Обрезка графика с помощью masked_where() и xlim')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

## Постройте stairs-график (ступенчатый) для функции y(x). При построении графиков вид функций и параметры выбрать согласно варианту. Толщину линии, размер и цвет маркера, а также контура маркера подобрать самостоятельно.

## Постройте stem-график для функции y(x), xε[0,2π], ∆x=0,2. При построении графиков вид функций и параметры выбрать согласно варианту. Толщину линии, размер и цвет маркера, а также контура маркера подобрать самостоятельно.


```python
x = np.arange(0, 2 * np.pi + 0.2, 0.2)
y = np.cos(x)

plt.figure(figsize=(8, 5))

markerline, stemlines, baseline = plt.stem(
    x, y,
    linefmt='pink',
    markerfmt='o',
    basefmt='k--',
)

plt.setp(markerline,
         markersize=7,
         markerfacecolor='pink',
         markeredgecolor='crimson',
         markeredgewidth=1.2)

plt.setp(stemlines,
         linewidth=2,
         linestyle='--',
         color='pink')

plt.title('Stem-график функции $y(x)=\cos(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

Исследуйте зависимость Y от (X1 – вариант 1, X2 – вариант 2) за 20 лет (Таблица 4.1) – постройте диаграмму рассеяния и сделайте вывод о наличии/отсутствии зависимости и ее характере.

Таблица:

| Пятилетка гг. | Средний Y | Средний X |
|--------------|-----------|-----------|
| 2004–2008    | 4 626     | 0.8366    |
| 2009–2013    | 4 296     | 0.6750    |
| 2014–2018    | 2 672     | 0.6220    |
| 2019–2023    | 1 678     | 0.4224    |
| Итого в среднем | 3 318  | 0.6390    |

Постройте столбчатые диаграммы: 1) для среднего Y по пятилеткам; 2) для среднего (X1 – вариант 1, X2 – вариант 2) по пятилеткам. Расположите две диаграммы рядом – по горизонтали.

Постройте круговую диаграмму, отражающую процентное соотношение Y по пятилеткам.

Сделайте общий вывод.

```python
df = pd.read_excel('table.xlsx')

# Оставляем только 4 пятилетки (без строки "итого в среднем")
df = df[df["пятилетка гг."].astype(str).str.contains(r"\d{4}-\d{4}", na=False)].copy()

def to_float(series: pd.Series) -> pd.Series:
    return (series.astype(str)
            .str.replace("\xa0", "", regex=False)  
            .str.replace(" ", "", regex=False)     
            .str.replace(",", ".", regex=False)   
            .astype(float))

df["Y"] = to_float(df["средний Y"])
df["X"] = to_float(df["Средний X"])

df
```

```python
# Диаграмма рассеяния Y от X 
plt.figure(figsize=(7, 5))
plt.scatter(df["X"], df["Y"])
plt.title("Диаграмма рассеяния Y от X за 20 лет")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

# Две столбчатые диаграммы рядом (по горизонтали)
fig, axes = plt.subplots(1, 2, figsize=(12, 4), constrained_layout=True)

axes[0].bar(df["пятилетка гг."], df["Y"])
axes[0].set_title("Средний Y по пятилеткам")
axes[0].set_xlabel("Пятилетка")
axes[0].set_ylabel("Y")
axes[0].grid(True, axis="y")
axes[0].tick_params(axis="x", rotation=20)

axes[1].bar(df["пятилетка гг."], df["X"])
axes[1].set_title("Средний X по пятилеткам")
axes[1].set_xlabel("Пятилетка")
axes[1].set_ylabel("X")
axes[1].grid(True, axis="y")
axes[1].tick_params(axis="x", rotation=20)

plt.show()
```

По результатам диаграммы рассеивания можно подметить: точки располагаются почти по возрастающей линии, при уменьшении X наблюдается существенное снижение Y, зависимость близкая к линейной. Между X и Y присутствует прямая зависимость: при увеличении X значение Y возрастает. Связь выражена достаточно чётко, что позволяет говорить о сильной положительной корреляции.

По столбчатой диаграмме для Y наблюдается максимальное значение — 2004–2008, далее наблюдается устойчивое снижение и минимум — 2019–2023. Для X наблюдается аналогичная динамика. Динамика X и Y по пятилеткам согласована, что дополнительно подтверждает наличие зависимости между переменными.

```python
# Круговая диаграмма: доли Y по пятилеткам
plt.figure(figsize=(7, 6))
plt.pie(
    df["Y"],
    labels=df["пятилетка гг."],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Процентное соотношение Y по пятилеткам")
plt.show()
```

Более 2/3 совокупного значения Y приходится на первые две пятилетки, что указывает на значительное снижение показателя Y в последние годы.

За рассматриваемые 20 лет между X и Y наблюдается устойчивая прямая зависимость. Оба показателя демонстрируют согласованное снижение во времени, а анализ по пятилеткам подтверждает изменение уровня Y с максимальными значениями в ранние периоды и минимальными — в последние годы.
